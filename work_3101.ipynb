```python
import numpy as np
```

```python
rng = np.random.RandomState(0)
```

```python
df = pd.DataFrame({'key':['A','B','C','A','B','C'],
                   'data1': range(6),
                   'data2': rng.randint(0,10,6)},
                   columns = ['key','data1','data2'])
```

```python
df
```

```python
df.groupby('key').transform(lambda x: x - x.mean())
```

```python
df
```

```python
def norm_by_data2(x):
  x['data1'] /= x['data2'].sum()
  return x
```

```python
df.groupby('key').apply(norm_by_data2)
```
